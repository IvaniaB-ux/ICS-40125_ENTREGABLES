```python
import pandas as pd

# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/netflix_titles.csv')
df.head()
```

**Parte 1: Limpieza y preparación**

```python
#Número de filas y columnas.

df.shape
```

```python
#Tipos de datos.

df.dtypes
```

```python
#Valores nulos por columna.

df.isnull().sum()
```

```python
#Convertir date_added a tipo fecha

df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
#Crear columnas con assign

df = df.assign(
    year_added=df["date_added"].dt.year,
    month_added=df["date_added"].dt.month
)
```

**Parte 2: Técnicas Avanzadas de pandas**

```python
#.loc

df_movies_2018 = df.loc[
    (df["type"] == "Movie") & (df["year_added"] > 2018)
]
```

```python
#str.contains() --> Títulos con "love"

df_love = df.loc[
    df["title"].str.contains("love", case=False, na=False)
]

#str.extract() --> duración (min)

df["duration_minutes"] = df["duration"].str.extract(r"(\d+)").astype(float)
```

```python
#explode()

df_exploded = df.explode("listed_in")
```

```python
#Top 10 géneros más frecuentes

top_genres = df_exploded["listed_in"].value_counts().head(10)
```

```python
#where() y mask()

df["is_long_where"] = df["duration_minutes"].where(
    df["duration_minutes"] > 120,
    other="No largo"
)

df["is_long_mask"] = df["duration_minutes"].mask(
    df["duration_minutes"] <= 120,
    other="No largo"
)
```

```python
#.loc

df_filtered = df.loc[
    (df["duration_minutes"] > 100) &
    (df["rating"] == "R") &
    (df["country"] == "United States")
]
```

**¿Cuáles son las combinaciones más frecuentes de género y rating en el dataset?**


Los resultados muestran que las combinaciones más frecuentes entre género y rating se concentran principalmente en contenido de tipo “Stand-Up Comedy” con clasificación “TV-MA”, siendo la combinación más repetida del dataset (284 apariciones).

En segundo lugar, destacan combinaciones de dramas e internacional movies, especialmente con ratings TV-MA y TV-14, lo que indica una fuerte presencia de contenido adulto o para audiencias mayores en producciones internacionales.

También aparecen con frecuencia combinaciones mixtas como:

Dramas + Independent + International Movies (TV-MA)
Comedias + Dramas + Internacionales (TV-14)

Esto sugiere que el catálogo de Netflix no está dominado por géneros “puros”, sino por contenidos híbridos, donde los dramas y las producciones internacionales tienen un rol central.

En categorías familiares o infantiles, aparecen combinaciones como:

Kids’ TV (TV-Y)
Children & Family Movies (PG)

pero con menor frecuencia en comparación con contenido adulto.
