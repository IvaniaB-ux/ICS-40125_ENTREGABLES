```python
import pandas as pd

# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT281/main/docs/labs/data/netflix_titles.csv')
df.head()
```


### Parte 1: Limpieza y preparación

1. Revisar y describir el dataset:

   * ¿Cuántas filas y columnas tiene?
   * ¿Qué tipos de datos hay?
   * ¿Cuántos valores nulos hay por columna?

2. Transformar la columna `date_added` a tipo fecha.

3. Crear columnas auxiliares con `assign`:

   * Año (`year_added`)
   * Mes (`month_added`)



```python
# Parte 1: Limpieza y preparación

# 1. Revisar dimensiones del dataset
print("Cantidad de filas y columnas:")
print(df.shape)

print("\nTipos de datos por columna:")
display(df.dtypes)

print("\nValores nulos por columna:")
display(df.isnull().sum())

# 2. Transformar date_added a tipo fecha
df["date_added"] = pd.to_datetime(df["date_added"].str.strip(), errors="coerce")

print("\nTipo de dato de date_added después de la transformación:")
print(df["date_added"].dtype)

# 3. Crear columnas auxiliares con assign
df = df.assign(
    year_added = df["date_added"].dt.year,
    month_added = df["date_added"].dt.month
)

# Vista previa del dataset preparado
df.head()
```

En esta primera etapa se revisó la estructura general del dataset de Netflix, identificando su cantidad de filas, columnas, tipos de datos y valores faltantes. Este paso es importante porque permite conocer la calidad inicial de la información antes de aplicar transformaciones.

Luego, la columna `date_added` fue convertida a formato de fecha, lo que permite trabajar con componentes temporales como año y mes. A partir de esta columna se crearon `year_added` y `month_added`, variables auxiliares útiles para analizar cuándo se agregaron los títulos al catálogo.

## Parte 2: Técnicas avanzadas de pandas

4. Utilizar `.loc` para seleccionar películas (`type == 'Movie'`) que fueron agregadas después del año 2018.

5. Utilizar `str.contains()` y `str.extract()`:

   * Filtrar títulos que contienen la palabra 'love' (sin distinguir mayúsculas/minúsculas).
   * Extraer la duración en minutos para las películas desde la columna `duration`.

6. Aplicar `explode()` sobre la columna `listed_in` para obtener una fila por cada género.

7. Obtener un top 10 de géneros más frecuentes utilizando `value_counts()`.

8. Aplicar `where()` y `mask()` para marcar las películas de más de 120 minutos como contenido largo en una nueva columna.

9. Utilizar `.loc` para filtrar películas que cumplen con:

   * Más de 100 minutos de duración.
   * Rating igual a `'R'`.
   * País igual a `'United States'`.

10. Utilizar `.style` para formatear visualmente el top 10 de películas más largas.

```python
# Parte 2: Técnicas avanzadas de pandas

# 4. Seleccionar películas agregadas después del año 2018 usando .loc
movies_after_2018 = df.loc[
    (df["type"] == "Movie") & (df["year_added"] > 2018)
]

print("Películas agregadas después de 2018:")
display(movies_after_2018[["title", "type", "date_added", "year_added"]].head())

# 5. Filtrar títulos que contienen la palabra 'love'
love_titles = df.loc[
    df["title"].str.contains("love", case=False, na=False)
]

print("\nTítulos que contienen la palabra 'love':")
display(love_titles[["title", "type", "release_year"]].head())

# Extraer duración en minutos para películas desde duration
df = df.assign(
    duration_minutes = df["duration"].str.extract(r"(\d+)").astype(float)
)

print("\nDuración extraída en minutos o temporadas:")
display(df[["title", "type", "duration", "duration_minutes"]].head())

# Para análisis de duración de películas, se filtra solo Movie
movies = df.loc[df["type"] == "Movie"].copy()

# 6. Aplicar explode sobre listed_in para obtener una fila por género
df_genres = df.assign(
    genre = df["listed_in"].str.split(", ")
).explode("genre")

print("\nDataset con una fila por género:")
display(df_genres[["title", "type", "genre", "rating"]].head(10))

# 7. Top 10 de géneros más frecuentes
top_10_genres = df_genres["genre"].value_counts().head(10)

print("\nTop 10 géneros más frecuentes:")
display(top_10_genres)

# 8. Usar where y mask para marcar películas de más de 120 minutos como contenido largo
movies = movies.assign(
    content_length = "Contenido normal"
)

movies["content_length"] = movies["content_length"].where(
    movies["duration_minutes"] <= 120,
    "Contenido largo"
)

# Alternativa con mask
movies["content_length_mask"] = "Contenido normal"
movies["content_length_mask"] = movies["content_length_mask"].mask(
    movies["duration_minutes"] > 120,
    "Contenido largo"
)

print("\nClasificación de duración de películas:")
display(movies[["title", "duration", "duration_minutes", "content_length", "content_length_mask"]].head())

# 9. Filtrar películas con más de 100 minutos, rating R y país United States
filtered_movies = movies.loc[
    (movies["duration_minutes"] > 100) &
    (movies["rating"] == "R") &
    (movies["country"].str.contains("United States", case=False, na=False))
]

print("\nPelículas con más de 100 minutos, rating R y país United States:")
display(filtered_movies[["title", "country", "rating", "duration_minutes"]].head(10))

# 10. Formatear visualmente el top 10 de películas más largas
top_10_longest_movies = movies.sort_values("duration_minutes", ascending=False).head(10)

top_10_longest_movies[["title", "country", "release_year", "rating", "duration_minutes"]].style.background_gradient(
    subset=["duration_minutes"],
    cmap="Blues"
).format({
    "duration_minutes": "{:.0f} min"
})
```

En esta sección se aplicaron distintas herramientas avanzadas de `pandas` para filtrar, transformar y reorganizar la información. Primero, se seleccionaron las películas agregadas después de 2018, lo que permite observar contenido incorporado más recientemente al catálogo.

También se filtraron títulos que contienen la palabra `love`, mostrando cómo trabajar con texto mediante `str.contains()`. Luego, se extrajo la duración numérica desde la columna `duration`, separando el valor cuantitativo del texto original. Este paso permite analizar la duración de las películas como una variable numérica.

Posteriormente, se utilizó `explode()` para separar los géneros listados en `listed_in`, generando una fila por cada género asociado a un título. Esto permite calcular de forma más clara cuáles son los géneros más frecuentes en Netflix. Además, se clasificaron las películas según su duración, identificando como contenido largo aquellas superiores a 120 minutos.

Finalmente, se aplicó un filtro más específico para identificar películas de más de 100 minutos, con rating `R` y producidas en Estados Unidos. El uso de `.style` permite presentar visualmente las películas más largas, facilitando la interpretación de los resultados.



### Pregunta Desafío

11. ¿Cuáles son las combinaciones más frecuentes de género y rating en el dataset?
    (Sugerencia: utilizar `value_counts` con `subset=["genre", "rating"]` después de aplicar `explode()`).



### Bonus: Análisis de duplicados y limpieza

12. ¿Existen películas con el mismo nombre (`title`) pero con distinto año de lanzamiento (`release_year`)?
13. ¿Cuántos títulos únicos hay en total en la columna `title`?





```python
# Pregunta desafío y bonus

# 11. Combinaciones más frecuentes de género y rating
genre_rating_combinations = (
    df_genres
    .value_counts(subset=["genre", "rating"])
    .head(10)
    .reset_index(name="frequency")
)

print("Top 10 combinaciones más frecuentes de género y rating:")
display(genre_rating_combinations)

# 12. Películas con el mismo título, pero distinto año de lanzamiento
title_year_unique = df[["title", "release_year"]].drop_duplicates()

same_title_different_year = title_year_unique.loc[
    title_year_unique["title"].duplicated(keep=False)
].sort_values(["title", "release_year"])

print("\nTítulos repetidos con distinto año de lanzamiento:")
display(same_title_different_year)

print("\nCantidad de títulos con el mismo nombre y distinto año:")
print(same_title_different_year["title"].nunique())

# 13. Cantidad de títulos únicos
unique_titles = df["title"].nunique()

print("\nCantidad total de títulos únicos:")
print(unique_titles)
```


Las combinaciones más frecuentes de género y rating permiten identificar qué tipos de contenido predominan en el catálogo según su clasificación etaria. Esta información puede ser útil para analizar la orientación del catálogo de Netflix hacia ciertos públicos y categorías de contenido.

El análisis de títulos repetidos permite detectar posibles remakes, contenidos con el mismo nombre o registros que podrían requerir revisión adicional. Para ello se compararon los títulos con sus años de lanzamiento, identificando aquellos casos en que un mismo nombre aparece asociado a más de un año.

Finalmente, la cantidad de títulos únicos entrega una medida más precisa del tamaño del catálogo, evitando contar repetidamente contenidos que puedan aparecer duplicados o con variaciones en alguna variable.

El análisis muestra que las combinaciones más frecuentes entre género y clasificación etaria se concentran principalmente en contenidos internacionales y clasificaciones para público adulto o adolescente. La combinación más repetida es **International Movies con clasificación TV-MA**, con **1130 títulos**, seguida por **International Movies con clasificación TV-14**, con **1065 títulos**. Esto indica que el catálogo de Netflix tiene una fuerte presencia de películas internacionales orientadas principalmente a audiencias adultas y adolescentes.

También destacan los **Dramas**, tanto con clasificación **TV-MA** como **TV-14**, lo que sugiere que este género tiene un peso importante dentro del catálogo. Además, aparecen categorías como **International TV Shows**, **Comedies** y **TV Dramas**, lo que evidencia una oferta diversa, pero concentrada en ciertos géneros de alta frecuencia.

Respecto al análisis de títulos repetidos con distinto año de lanzamiento, el resultado fue **0**. Esto significa que, dentro del dataset, no se encontraron títulos que tengan exactamente el mismo nombre pero distinto año de estreno. Por lo tanto, no se identifican posibles remakes, registros duplicados o títulos repetidos bajo distintos años de lanzamiento según el criterio utilizado.

Finalmente, la cantidad total de títulos únicos fue de **8807**. Este resultado indica que el catálogo analizado contiene 8807 títulos distintos.
