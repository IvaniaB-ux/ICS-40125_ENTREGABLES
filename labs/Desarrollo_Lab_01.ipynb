**Problema 1.  Aproximando  π  con una serie infinita**

```python
def calcular_pi(n):
  suma = 0

  for k in range(1, n + 1):
    termino = ((-1)**(k+1))/(2*k - 1)
    suma+= termino

  pi = 4 * suma
  return pi
```

**Ejecución de prueba:**



calcular_pi(3)

calcular_pi(1000)

```python
print(calcular_pi(3))
print(calcular_pi(1000))
```

**Problema 2. Aproximando  e  con la serie de Euler**

Parte A: Función factorial(n)

```python
def factorial(n):
    resultado = 1

    for i in range(1, n + 1):
        resultado *= i

    return resultado
```

Parte B: Función calcular_e(n)

```python
def calcular_e(n):
    suma = 0

    for k in range(n+1):
        suma += 1 / factorial(k)

    return suma
```

**Ejecución de prueba:**



factorial(3)

factorial(5)

calcular_e(3)

calcular_e(1000)

```python
print(factorial(3))
print(factorial(5))

print(calcular_e(3))
print(calcular_e(1000))
```

**Problema 3. Números amigos y suma de divisores propios**

Parte A: Función divisores_propios(n)

```python
def divisores_propios(n):
    divisores = []

    for i in range(1, n):
        if n % i == 0:
            divisores.append(i)

    return divisores
```

Parte B: Función amigos(n1, n2)

```python
def amigos(n1, n2):

    suma1 = sum(divisores_propios(n1))
    suma2 = sum(divisores_propios(n2))

    return suma1 == n2 and suma2 == n1
```

**Ejecución de prueba:**


divisores_propios (220)

divisores_propios(284)

amigos(220,284)

amigos (6,5)




```python
print(divisores_propios (220))

print(divisores_propios(284))

print(amigos(220,284))

print(amigos (6,5))
```

**Problema 4. La conjetura de Collatz**

```python
def collatz(n):

    secuencia = [n]

    while n != 1:

        if n % 2 == 0:
            n = n // 2

        else:
            n = 3 * n + 1

        secuencia.append(n)

    return secuencia
```

**Ejecución de prueba:**

collatz (9)

```python
print(collatz(9))
```

**Problema 5. La conjetura de Goldbach**

Parte A. es_primo(n)

```python
def es_primo(n):

    if n < 2:
        return False

    for i in range(2, n):

        if n % i == 0:
            return False

    return True
```

Parte B. lista_de_primos(n)

```python
def lista_de_primos(n):

    primos = []

    for i in range(2, n + 1):

        if es_primo(i):
            primos.append(i)

    return primos
```

Parte C. goldbach(n)

```python
def goldbach(n):

    primos = lista_de_primos(n)

    for p1 in primos:

        for p2 in primos:

            if p1 + p2 == n:
                return p1, p2
```
